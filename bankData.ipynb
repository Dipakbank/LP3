```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
bank_data = pd.read_csv("/content/sample_data/bank/bank_marketing.csv")
```

```python
bank_data.dropna(inplace=True)

```

```python
bank_data.head()
```

```python
# Drop unnecessary columns, missing values, and duplicates

bank_data.drop(columns=['Unnamed: 0'], inplace=True)
bank_data.dropna(inplace=True)
bank_data.drop_duplicates(inplace=True)
```

```python

# Assuming 'df' is your DataFrame containing the data
categorical_columns = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome']
bank_data_categorical = pd.get_dummies(bank_data, columns=categorical_columns)
```

```python
bank_data_categorical.head()
```

```python
from sklearn.model_selection import train_test_split

# Assuming 'bank_data_categorical' is your DataFrame
X = bank_data_categorical.drop('deposit', axis=1)  # Independent variables
y = bank_data_categorical['deposit']  # Target variable

# Split the data into training (70%) and testing (30%) sets with a random state of 0
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
k = 11
knn_model = KNeighborsClassifier(n_neighbors=k)
knn_model.fit(X_train, y_train)
```

```python
y_pred = knn_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy of the model:", accuracy)
```

```python
misclassified_samples = (y_test != y_pred).sum()
print("Number of misclassified samples:", misclassified_samples)
```
